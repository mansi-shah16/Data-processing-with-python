```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm

example_file = ("C:/Users/Lenovo/Downloads/q3.jpg")
image = imread(example_file, as_gray = True)
image3 = resize(image,(70,60), mode = 'symmetric')

plt.imshow(image3, cmap = cm.gray)
print("data type: %s, shape: %s"%(type(image3), image3.shape))
```

```python
image_row=image3.flatten()
print ("data type:%s,shape:%s"%(type(image_row),image_row.shape))
```

```python
#P4_8
from lxml import objectify
import pandas as pd
xml = objectify.parse("XMLData.xml")
root = xml.getroot()
df = pd.DataFrame(columns=('Number', 'String', 'Boolean'))
for i in range(0,4):
    obj = root.getchildren()[i].getchildren()
    row = dict(zip(['Number', 'String', 'Boolean'],[obj[0].text,
    obj[1].text,obj[2].text]))
    row_s = pd.Series(row)
    row_s.name = i
    df = df.append(row_s)
print(df)
```
